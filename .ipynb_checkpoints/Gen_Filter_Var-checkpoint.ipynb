## Generate Unique For Each Feature Variable

```python
def df_unique(arr, feat_name):
    tmp_keys = []
    for val in arr:
        if type(val) != str:
            continue
        tmp_list = val.split(";")
        for val2 in tmp_list:
            if val2 not in tmp_keys:
                tmp_keys.append(val2)
                
    arr_list = []
    for val in arr:
        if type(val) != str:
            continue
        tmp_dict = {i:"No" for i in tmp_keys}
        tmp_list = val.split(";")
        for val2 in tmp_list:
            tmp_dict[val2] = "Yes"
        arr_list.append(tmp_dict)
        
    tmp_df = pd.DataFrame(arr_list)
    tmp_df.columns = [f"[{feat_name}-{val}]" for val in tmp_df.columns]
    
    return tmp_df
```
